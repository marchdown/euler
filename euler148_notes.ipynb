Divisors (and non-divisors) of 7 (as with any prime number $p$) 
for a repeating self-similar pattern with cells of size $p^n$ for a $n \in \mathbb{N}$

The cells behave as follows: 
the top 7 rows represent all the possible remainders and the 7th line flips all inner elements to zero

```haskell
mapM_ print [pascalsTriangleLine n | n <- [1..7]]
```

```haskell
mapM_ print [map (flip rem 7) $ pascalsTriangleLine n | n <- [1..7]]
```
